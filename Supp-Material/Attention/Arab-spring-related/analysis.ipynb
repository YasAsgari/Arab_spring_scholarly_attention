```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns

import matplotlib as mpl


mpl.rcParams.update({
    'font.family':          'sans-serif',
    'font.sans-serif':      ['Arial'],
    'mathtext.fontset':     'custom',
    'mathtext.rm':          'Arial',
    'mathtext.it':          'Arial:italic',
    'mathtext.bf':          'Arial:bold',
    'mathtext.cal':         'Arial',        # for calligraphic if you like
})
```

```python
df=pd.read_csv('Scopus-arabspring.csv')
```

```python
counts=df.groupby('Year').count()['Title'].reset_index().rename(columns={'Title':'Counts'})
```

```python
fig, ax = plt.subplots(figsize=(6,4), dpi=500)
ax.bar(counts['Year'], counts['Counts'], color="#9ecae1")
ax.plot(counts['Year'], counts['Counts'], color="#08519c", marker='o')

ax.set_xlabel('Year')
ax.set_ylabel('Number of Articles')
ax.set_xticks(counts['Year'])
plt.xticks(rotation=90);
ax.axvline(2011, color='k', linestyle='--')
fig.savefig('arab_spring_publications_over_years_final.pdf', dpi=400, bbox_inches='tight')
```
