```python
import numpy as np
import pandas as pd
```

```python
df1=pd.read_csv(r"C:\Users\Yasaman\Downloads\samples_1127.xlsx - Annotated by Hongyu.csv")
df2=pd.read_csv(r"C:\Users\Yasaman\Downloads\samples_1127.xlsx - Annotated by Yas.csv")
```

```python
df2['code']=df2['Unnamed: 9']
```

```python
res=df1.merge(df2, on=['ID', 'Year', 'Field', 'SampleGroup', 'Title', 'All_text', 'Abstract'], suffixes=['_hongyu', '_yas'], how='outer')
res=res.drop(columns=['Unnamed: 9', 'All_text','RandomOrder_yas', 'RandomOrder_hongyu'])
res['code_yas']=res['code_yas'].apply(lambda x: sorted(x.lower().replace(' ', '').split(';')))
res['code_hongyu']=res['code_hongyu'].apply(lambda x: sorted(x.lower().replace(' ', '').split(';')))
```

```python
res.to_csv('res_new.csv', index=False)
```
