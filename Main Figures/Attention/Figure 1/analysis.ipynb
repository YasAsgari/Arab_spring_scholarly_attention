```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
df = pd.read_csv(r"C:\Users\Yasaman\Downloads\Attention-fractional counting.csv")
df=df[(df['year']>=2002)&(df['year']<=2019)]
```

```python
x=df[(df['country']=='egy')&(df['affiliation_country']=='sau')].groupby(['year','subjarea'] ).sum().reset_index()#[['year', 'aggregated_value']].sort_values(by='year' )
```

```python

plt.figure(figsize=(10,6))
plt.bar(x['year'], x['aggregated_value'], color='skyblue')
plt.xlabel('Year')
plt.ylabel('Aggregated Attention Value')

plt.xticks(x['year'], rotation=45)
plt.title('Aggregated Attention from Saudi Arabia to Egypt Over the Years')
```
